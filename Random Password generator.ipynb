# Random Password Generator

```python
import string
import random
```

```python
# Minimum 8 latters 
# digit ,uppercase and symbol

password = random.choice(string.ascii_uppercase) + random.choice(string.digits) + random.choice(string.punctuation)
n = int(input("Enter length of your password..."))
i = 1
while i <= n - 3:
    password += random.choice(string.ascii_letters + string.punctuation + string.digits)
    i += 1

else:
    print(f"Your Random Password is {password}")
    print(f"Length of password is {len(password)}")

# you have to enter the number how long password you want
```
